```python
def my_thermo_stat(temp, desired_temp):
    if temp < desired_temp - 5:
        status = "Heat"
    elif temp > desired_temp + 5:
        status = "AC"
    else:
      status = "Off"
    return status
```

```python
status = my_thermo_stat(65, 75)
print(status)
```

```python
status = my_thermo_stat(75, 65)
print(status)
```

```python
status = my_thermo_stat(65, 63)
print(status)
```
